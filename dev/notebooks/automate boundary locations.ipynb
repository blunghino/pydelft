```python
import numpy as np
from pydelft import grd, dep, bnd
import matplotlib.pyplot as plt
import matplotlib as mpl
% matplotlib inline
```

```python
fname = r'z:\Aleutians\models\sedanka_2d_profiles\eq_sources\nesting\50m_slip\stardust\sedanka_bathy_asp_5m_wgs84.dep'
depth = dep.dep()
depth.read_dep(fname)
```

```python
fname = r'z:\Aleutians\models\sedanka_2d_profiles\eq_sources\nesting\50m_slip\stardust\stardust_asp_wgs84_5m.grd'
grid = grd.grd()
grid.read_grd(fname)
```

```python
fig = plt.figure(figsize = (15,8))
plt.pcolormesh(grid.x,grid.y,np.ma.masked_less(-depth.depth,0), vmin = 0, vmax = 50, cmap = mpl.cm.YlGn_r)
plt.pcolormesh(grid.x,grid.y, np.ma.masked_greater_equal(-depth.depth,0), vmin = -50, vmax = 0, cmap = mpl.cm.Blues_r)

for i in np.arange(1190,1728,50):
    plt.plot(grid.x[0,i], grid.y[0,i], 'ro', alpha = 0.5)
plt.plot(grid.x[0,1728], grid.y[0,1728],'ro', alpha = 0.5)

a = np.array(np.where(-depth.depth[:,-2] <= 0))
for i in np.arange(np.min(a), np.max(a), 50):
    plt.plot(grid.x[i,grid.n-1], grid.y[i,grid.n-1],'yo', alpha = 0.5)
plt.plot(grid.x[np.max(a),grid.n-1], grid.y[np.max(a),grid.n-1],'yo', alpha = 0.5)
```
