<h3>Find the m,n index of the grid that corresponds as close as possible to the station coordinates</h3>
We can write a simple function that takes a coordinate pair, subtracts it from all the grid lons and lats, and returns where the difference is a minimum.

```python
def find_nearest(grid, query):
    m = np.unravel_index(np.abs(grid.y-query[1]).argmin(), np.shape(grid.y))[0]
    n = np.unravel_index(np.abs(grid.x-query[0]).argmin(), np.shape(grid.x))[1]
    return [m,n]            
```

Test out the function, print the indices and the lon/lat on the grid that are at those indices

```python
idx = find_nearest(grid, [lon,lat])
print('indices: %s' % idx)
print(grid.x[idx[0],idx[1]], grid.y[idx[0],idx[1]])
```

Let's convert all the sites now

```python
obs_idx = [find_nearest(grid,[i[1],i[2]]) for i in sites_coords]
```

<h3>Plot on grid</h3>

```python
depth = dep(r'i:\Japan\Data\Sendai\T3\create_riemann\testing\run6\natori_10_v1.0.dep', grid_fname = grid.filename)
ll_lon, ur_lon = np.min(grid.x), np.max(grid.x)
ll_lat, ur_lat = np.min(grid.y), np.max(grid.y)
m = Basemap(projection='tmerc',llcrnrlon=ll_lon, llcrnrlat = ll_lat, 
            urcrnrlon=ur_lon, urcrnrlat=ur_lat, resolution='i', 
            lon_0=ll_lon, lat_0 = ll_lat)
x, y = m(grid.x, grid.y)
```

```python
fig = plt.figure(figsize=(12,18))
ax = fig.add_subplot(1,1,1)
m.contour(x,y,-depth.depth, np.arange(0,100,1), colors = 'k')
for i in obs_idx:
    obs_point = m.plot(x[i[0],i[1]],y[i[0],i[1]], 'bo',
                       label = 'point on grid')
for i in sites_coords:
    sx,sy = m(i[1],i[2])
    site_point = m.plot(sx,sy, 'r*',
                        label = 'actual location')

handles, labels = ax.get_legend_handles_labels()
legend = ax.legend([handles[0], handles[-1]], [labels[0],labels[-1]])
```

<h3>Write to .obs file</h3>

```python
f = open(r'..\test_data\natori_10_v1.obs','w')
num_obs = np.shape(obs_idx)[0]
for i in range(0,num_obs-1):
    name = sites_coords[i][0].ljust(20)
    line = str('%s\t%s\t%s\n' %(name,int(obs_idx[i][1]), 
                                int(obs_idx[i][0])))
    if len(line) > 132:
        print('ERROR: record length too long, max 132\nAt %s' 
              % line)
        break
    f.write(line)
f.close()
```
